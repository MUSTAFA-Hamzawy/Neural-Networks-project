```python
import joblib
import cv2
from feature_extraction import extract_features
from preprocessing import shadow_removalDialte, preprocessing
```

```python
model = joblib.load('model/xgb_model.pkl')
```

```python
# Define the path to the text file
output_file = "./result.txt"

# Open the text file in write mode
with open(output_file, "w") as file:

    for i in range(1, 49):
        im = cv2.imread(f'./test/({i}).jpg')
        img = cv2.resize(im, (256, 256))

        # pre-processing
        result_image = shadow_removalDialte(img)
        preprocessed_image = preprocessing(result_image, img)

        # feature extraction
        fd = extract_features(preprocessed_image)

        pred = model.predict([fd])
        pred = int(pred[0])
        file.write(str(pred) + "\n")  # Write the predicted number to the file followed by a new line
```

```python
correct_file =  "./results_set_1.txt" # Path to the file with correct answers
predicted_file = "./result.txt"  # Path to the file with predicted results

correct = 0
# Read the correct file
with open(correct_file, "r") as correct_f:
    correct_lines = correct_f.readlines()

# Read the predicted file
with open(predicted_file, "r") as predicted_f:
    predicted_lines = predicted_f.readlines()

# Compare the values line by line
for i in range(len(correct_lines)):
    correct_value = int(correct_lines[i].strip())
    predicted_value = int(predicted_lines[i].strip())

    if correct_value == predicted_value:
        correct += 1
        print(f"Prediction {i+1}: Correct")
    else:
        print(f"Prediction {i+1}: Incorrect")
        print(f"Correct value: {correct_value}")
        print(f"Predicted value: {predicted_value}")
    print("--------------------")
print("accuarcy: ", correct/len(correct_lines)*100, "%")
```
